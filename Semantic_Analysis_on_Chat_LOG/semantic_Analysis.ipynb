```python
from textblob import TextBlob

def analyze_sentiment(sentence):
    # Create a TextBlob object
    blob = TextBlob(sentence)
    
    # Perform sentiment analysis
    sentiment = blob.sentiment
    
    # Print the sentiment polarity and subjectivity
    print(f"Sentiment Polarity: {sentiment.polarity}")
    print(f"Sentiment Subjectivity: {sentiment.subjectivity}")
    
    # Determine sentiment polarity
    if sentiment.polarity > 0:
        print("Overall Sentiment: Positive")
    elif sentiment.polarity < 0:
        print("Overall Sentiment: Negative")
    else:
        print("Overall Sentiment: Neutral")

# Example sentence
sentence = "Hi, I am really frustrated right now. I bought a new laptop from your store last week, and it is already not working. It will not turn on at all."

# Analyze sentiment
analyze_sentiment(sentence)
```

```python
from textblob import TextBlob

# Define the chat log with dialogues and speakers
chat_log = [
    {"speaker": "Customer Support", "text": "Hi there! Thank you for contacting ABC Electronics. How can I help you today?"},
    {"speaker": "Customer", "text": "Hi, I’m really frustrated right now. I bought a new laptop from your store last week, and it's already not working. It won't turn on at all."},
    {"speaker": "Customer Support", "text": "I'm really sorry to hear that you're having trouble with your new laptop. Let's see what we can do to get this resolved for you. Can you please provide me with your order number?"},
    {"speaker": "Customer", "text": "Sure, it's #123456."},
    {"speaker": "Customer Support", "text": "Thank you. Let me check the details of your order. Please hold on for a moment."},
    {"speaker": "Customer", "text": "Alright."},
    {"speaker": "Customer Support", "text": "Thank you for waiting. I see that your laptop is still under warranty. Let’s try a few troubleshooting steps to see if we can get it working. Can you please check if the power cable is properly connected and if the power indicator light is on?"},
    {"speaker": "Customer", "text": "Yes, I've checked that multiple times. The cable is connected, but there's no light at all."},
    {"speaker": "Customer Support", "text": "I understand how frustrating that must be. It sounds like it could be an issue with the power supply. Have you tried using a different outlet or a different power cable, if you have one available?"},
    {"speaker": "Customer", "text": "Yes, I've tried different outlets and even a different cable. Nothing works. I’m really worried now. I need this laptop for my work."},
    {"speaker": "Customer Support", "text": "I completely understand your concern. Since the troubleshooting steps didn't work, we can arrange for a replacement for you. Would you prefer a replacement or a refund?"},
    {"speaker": "Customer", "text": "A replacement would be fine. How long will it take?"},
    {"speaker": "Customer Support", "text": "Typically, it takes 3-5 business days to process a replacement and ship it to you. We’ll also provide you with a prepaid shipping label to return the defective laptop."},
    {"speaker": "Customer", "text": "That’s a bit long, but I guess I don’t have much choice. I really hope the new one works. This whole situation is so stressful."},
    {"speaker": "Customer Support", "text": "I can imagine how stressful this must be. We’ll do everything we can to expedite the process for you. In the meantime, is there anything else I can assist you with?"},
    {"speaker": "Customer", "text": "No, that’s all for now. Just get that replacement to me as soon as possible."},
    {"speaker": "Customer Support", "text": "Absolutely. I’ve initiated the replacement process, and you should receive an email with further instructions shortly. If you have any questions or need further assistance, please don’t hesitate to contact us again."},
    {"speaker": "Customer", "text": "Thank you. I’m feeling a bit better knowing this will be taken care of."},
    {"speaker": "Customer Support", "text": "You’re welcome. We appreciate your patience and understanding. Have a great day!"},
    {"speaker": "Customer", "text": "You too."}
]

def analyze_dialogue_sentiment(dialogues):
    sentiments = []
    for dialogue in dialogues:
        # Perform sentiment analysis for each dialogue
        blob = TextBlob(dialogue["text"])
        sentiment = blob.sentiment.polarity
        sentiments.append(sentiment)
        
        # Print dialogue and sentiment analysis results
        print(f"{dialogue['speaker']} - Dialogue: {dialogue['text']}")
        print(f"Sentiment Polarity: {sentiment:.2f}")
        if sentiment > 0:
            print("Positive")
        elif sentiment < 0:
            print("Negative")
        else:
            print("Neutral")
        print("---")
    
    # Calculate overall sentiment for the speaker
    avg_sentiment = sum(sentiments) / len(sentiments)
    print(f"Average Sentiment for {dialogues[0]['speaker']}: {avg_sentiment:.2f}")
    print("")

# Separate dialogues for Customer and Customer Support
customer_dialogues = [dialogue for dialogue in chat_log if dialogue["speaker"] == "Customer"]
customer_support_dialogues = [dialogue for dialogue in chat_log if dialogue["speaker"] == "Customer Support"]

# Analyze sentiment for Customer's dialogues
print("Sentiment Analysis for Customer:")
analyze_dialogue_sentiment(customer_dialogues)

# Analyze sentiment for Customer Support's dialogues
print("Sentiment Analysis for Customer Support:")
analyze_dialogue_sentiment(customer_support_dialogues)

# Calculate overall sentiment for the entire conversation
all_dialogues = [dialogue["text"] for dialogue in chat_log]
blob_all = TextBlob(" ".join(all_dialogues))
overall_sentiment = blob_all.sentiment.polarity

print("Overall Sentiment for the Entire Conversation:")
print(f"Overall Sentiment Polarity: {overall_sentiment:.2f}")
```
